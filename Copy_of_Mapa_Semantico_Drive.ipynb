```python
import os
import pandas as pd
from sentence_transformers import SentenceTransformer
from pypdf import PdfReader
import docx

# Nos aseguramos de que no hayamos perdido la variable
if 'archivos' not in locals():
    archivos = [f for f in os.listdir(RUTA_CARPETA) if os.path.isfile(os.path.join(RUTA_CARPETA, f))]

# Cargamos el modelo
model = SentenceTransformer('all-MiniLM-L6-v2')
data = []

print(f"Iniciando el análisis de archivos en '{RUTA_CARPETA}'...")

for nombre in archivos:
    ruta_completa = os.path.join(RUTA_CARPETA, nombre)
    texto = ""

    try:
        # 1. Leer PDFs
        if nombre.lower().endswith('.pdf'):
            reader = PdfReader(ruta_completa)
            for page in reader.pages[:5]: # Primeras 5 páginas
                texto += page.extract_text()

        # 2. Leer TXT o Markdown
        elif nombre.lower().endswith('.txt') or nombre.lower().endswith('.md'):
            with open(ruta_completa, 'r', encoding='utf-8') as f:
                texto = f.read()[:3000] # Primeros 3000 caracteres

        # 3. Leer Word (.docx)
        elif nombre.lower().endswith('.docx'):
            doc = docx.Document(ruta_completa)
            texto = "\n".join([para.text for para in doc.paragraphs])[:3000]

        # 4. Archivos nativos de Google Docs (.gdoc, .gsheet, etc)
        elif nombre.lower().endswith('.gdoc') or nombre.lower().endswith('.gsheet'):
            print(f"⏭️ Saltando: '{nombre}' (Es un enlace web de Google, no un archivo de texto local)")
            continue

        else:
            continue # Si es una imagen, zip u otro formato, lo ignoramos

        # Si logramos extraer texto, creamos el vector
        if len(texto.strip()) > 20:
            embedding = model.encode(texto)
            data.append({'nombre': nombre, 'embedding': embedding})
            print(f"✅ Procesado: {nombre[:40]}...")

    except Exception as e:
        print(f"⚠️ Error leyendo '{nombre}': {e}")

df = pd.DataFrame(data)
print(f"\n✨ ¡Magia terminada! Tenemos {len(df)} archivos listos para mapear.")
```

```python
import os
import pandas as pd
from sentence_transformers import SentenceTransformer
from pypdf import PdfReader
import docx

# --- AQUÍ ESTABA EL TRUCO: El espacio al final ---
RUTA_CARPETA = '/content/drive/MyDrive/Todos mis Docs y TXT '
# ------------------------------------------------

print(f"Iniciando el análisis en la carpeta: '{RUTA_CARPETA}'...")

# Listamos los archivos de la carpeta
archivos = [f for f in os.listdir(RUTA_CARPETA) if os.path.isfile(os.path.join(RUTA_CARPETA, f))]
print(f"Se encontraron {len(archivos)} archivos. Cargando el modelo de IA...\n")

# Cargamos el modelo
model = SentenceTransformer('all-MiniLM-L6-v2')
data = []

for nombre in archivos:
    ruta_completa = os.path.join(RUTA_CARPETA, nombre)
    texto = ""

    try:
        # 1. Leer PDFs
        if nombre.lower().endswith('.pdf'):
            reader = PdfReader(ruta_completa)
            for page in reader.pages[:5]: # Leemos las 5 primeras páginas
                if page.extract_text():
                    texto += page.extract_text()

        # 2. Leer TXT o Markdown
        elif nombre.lower().endswith('.txt') or nombre.lower().endswith('.md'):
            with open(ruta_completa, 'r', encoding='utf-8') as f:
                texto = f.read()[:3000] # Primeros 3000 caracteres

        # 3. Leer Word (.docx)
        elif nombre.lower().endswith('.docx'):
            doc = docx.Document(ruta_completa)
            texto = "\n".join([para.text for para in doc.paragraphs])[:3000]

        # 4. Archivos nativos de Google Docs
        elif nombre.lower().endswith('.gdoc') or nombre.lower().endswith('.gsheet'):
            print(f"⏭️ Saltando (Google Doc nativo): '{nombre}'")
            continue

        else:
            print(f"⏭️ Saltando (Formato no analizable): '{nombre}'")
            continue

        # Convertir a vector si hay texto
        if len(texto.strip()) > 20:
            embedding = model.encode(texto)
            data.append({'nombre': nombre, 'embedding': embedding})
            print(f"✅ Vectorizado: {nombre[:40]}...")

    except Exception as e:
        print(f"⚠️ Error leyendo '{nombre}': {e}")

# Guardamos los resultados
df = pd.DataFrame(data)
print(f"\n✨ ¡Magia terminada! Tenemos vectores listos para {len(df)} archivos.")
```

```python
import umap
import plotly.express as px
import numpy as np

print("Calculando el mapa (UMAP)...")

# Convertimos la columna de embeddings en una matriz matemática
matrix = np.vstack(df['embedding'].values)

# UMAP: Comprime nuestros vectores de muchas dimensiones a solo 2 (X, Y)
# Ajustamos los parámetros para 153 puntos
reducer = umap.UMAP(n_neighbors=5, min_dist=0.3, random_state=42)
embedding_2d = reducer.fit_transform(matrix)

# Añadimos las coordenadas X e Y a nuestro DataFrame
df['x'] = embedding_2d[:, 0]
df['y'] = embedding_2d[:, 1]

# Extraemos la extensión del archivo para pintarlos de colores distintos
df['tipo'] = df['nombre'].apply(lambda x: x.split('.')[-1] if '.' in x else 'desconocido')

print("¡Generando el gráfico interactivo!")

# Creamos el gráfico con Plotly
fig = px.scatter(
    df,
    x='x',
    y='y',
    hover_name='nombre',
    color='tipo', # Color según si es .docx, .txt, etc.
    title='Mapa Semántico de mis Archivos (153 Documentos)',
    template='plotly_dark', # Tema oscuro estilo hacker 😎
    color_discrete_sequence=px.colors.qualitative.Pastel
)

fig.update_traces(marker=dict(size=12, opacity=0.8, line=dict(width=1, color='White')))
fig.update_layout(showlegend=True, hoverlabel=dict(font_size=14))

# Mostramos la gráfica
fig.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
import plotly.express as px

print("Agrupando los archivos en 6 categorías principales (Clusters)...")

# 1. Agrupar con K-Means (usamos la matriz original de alta dimensión para mayor precisión)
numero_de_grupos = 6
kmeans = KMeans(n_clusters=numero_de_grupos, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(matrix)

print("Extrayendo los temas principales de cada grupo...\n")

# 2. Encontrar de qué trata cada grupo leyendo sus palabras
nombres_clusters = {}

for i in range(numero_de_grupos):
    # Obtenemos todos los archivos que cayeron en este cluster
    archivos_del_cluster = df[df['cluster'] == i]

    # Vamos a recopilar un poco de texto de cada archivo para analizarlo
    textos_combinados = []
    for nombre in archivos_del_cluster['nombre']:
        ruta_completa = os.path.join(RUTA_CARPETA, nombre)
        try:
            if nombre.lower().endswith('.txt') or nombre.lower().endswith('.md'):
                with open(ruta_completa, 'r', encoding='utf-8') as f:
                    textos_combinados.append(f.read()[:1000])
            elif nombre.lower().endswith('.docx'):
                doc = docx.Document(ruta_completa)
                textos_combinados.append("\n".join([para.text for para in doc.paragraphs])[:1000])
            elif nombre.lower().endswith('.pdf'):
                reader = PdfReader(ruta_completa)
                if len(reader.pages) > 0:
                     textos_combinados.append(reader.pages[0].extract_text()[:1000])
        except:
            pass # Si no podemos leer uno rápido, lo saltamos

    # Analizamos las palabras clave (ignoramos palabras vacías como "el", "la", "and", "the")
    # Usamos stopwords en español, inglés e italiano por los tipos de archivos que vimos
    if textos_combinados:
        vectorizer = TfidfVectorizer(max_features=5, stop_words=['el', 'la', 'los', 'las', 'un', 'una', 'y', 'o', 'en', 'de', 'que', 'a', 'por', 'para', 'con', 'and', 'the', 'to', 'of', 'in', 'is', 'it', 'for', 'il', 'lo', 'la', 'i', 'gli', 'le', 'di', 'a', 'da', 'in', 'con', 'su', 'per', 'tra', 'fra', 'non', 'che', 'si', 'è'])
        try:
            tfidf_matrix = vectorizer.fit_transform(textos_combinados)
            palabras_clave = vectorizer.get_feature_names_out()
            # La etiqueta será "Grupo X: palabra1, palabra2..."
            etiqueta = f"Grupo {i}: " + ", ".join(palabras_clave)
        except ValueError:
            etiqueta = f"Grupo {i}: (Textos muy cortos o irreconocibles)"
    else:
        etiqueta = f"Grupo {i}: (Sin texto extraíble)"

    nombres_clusters[i] = etiqueta
    print(f"👉 {etiqueta} ({len(archivos_del_cluster)} archivos)")

# 3. Aplicamos los nuevos nombres al DataFrame
df['nombre_cluster'] = df['cluster'].map(nombres_clusters)

print("\n¡Dibujando el mapa actualizado!")

# 4. Creamos el gráfico nuevo, coloreado por los temas encontrados
fig = px.scatter(
    df,
    x='x',
    y='y',
    hover_name='nombre',
    color='nombre_cluster', # ¡AQUÍ ESTÁ LA MAGIA! Color por tema.
    title='Mapa Semántico Agrupado por Temas',
    template='plotly_dark',
    color_discrete_sequence=px.colors.qualitative.Set1
)

fig.update_traces(marker=dict(size=12, opacity=0.8, line=dict(width=1, color='White')))

# Ajustamos la leyenda para que se lea bien (puede ser larga)
fig.update_layout(
    showlegend=True,
    hoverlabel=dict(font_size=14),
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=-0.3,
        xanchor="center",
        x=0.5
    )
)

fig.show()
```
